## Dealing With Outliers

```python
# Checking for outliers
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib.pyplot as plt
s_col=data.select_dtypes(include=['int','float'])
for i in s_col:
    sns.boxplot(data[i])
    plt.show()
```

```python
# Treating Outliers
skew_col=['symboling','normalized-losses','wheel-base','width','engine-size','horsepower','compression-ratio','stroke','peak-rpm','city-mpg','highway-mpg']
for col in skew_col:
    q25=data[col].quantile(0.25)
    q75=data[col].quantile(0.75)
    upper=q75+1.5*(q75-q25)
    lower=q25-1.5*(q75-q25)
    data[col]=np.where(data[col]>upper,upper,np.where(data[col]<lower,lower,data[col]))
    sns.boxplot(data[col])
    plt.show()
```

```python
data.describe()
```

```python
# Skewness (Before log_Transformation)
from scipy.stats import skew
for col in s_col:
    print(f'{col} Skewness:  {skew(data[col])}')
```

```python
for col in s_col:
    sns.distplot(data[col]) #(hist=False,kde=False)
    plt.show()
```

```python
# Treating skewness with log transformation
for col in s_col:
    if (data[col]<=0).any():
        print(f'{col} skip as it is either zero or negative value')
        continue
    data[col]=np.log(data[col])
    sns.distplot(data[col])
    plt.title(f'log_transformation : {col}')
    plt.show()
    print(f'{col} Skewness:  {skew(data[col])}')
```

```python
# (After log_Transformation)
for col in s_col:
    print(f'{col} Skewness:  {skew(data[col])}')
```

```python
# reseting index
for i in [x_train,y_train,x_val,y_val,x_test,y_test]:
    data.reset_index(drop=True,inplace=True)
```

### Models to test

* GradientBoostingRegressor
* RandomForestRegressor
* AdaBoostRegressor
* DecisionTreeRegressor
* LinearRegression
* SVR

```python
# Testing above listed models with default parameters
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor, AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold,GridSearchCV
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.pipeline import Pipeline

ss=StandardScaler()
kf=KFold(n_splits=5)
dtr= DecisionTreeRegressor()
gbr=GradientBoostingRegressor()
rfr=RandomForestRegressor()
abr=AdaBoostRegressor()
svr=SVR()

x_train=ss.fit_transform(x_train)
x_val=ss.transform(x_val)
x_test=ss.transform(x_test)
```

```python
gbr.fit(x_train,y_train)
y_preds=gbr.predict(x_val)
print('R2 Score : ',r2_score(y_val,y_preds))
print('MSE :      ',mean_squared_error(y_val,y_preds))
print('MAE :      ',mean_absolute_error(y_val,y_preds))
```

```python
abr.fit(x_train,y_train)
y_preds=abr.predict(x_val)
print('R2 Score : ',r2_score(y_val,y_preds))
print('MSE :      ',mean_squared_error(y_val,y_preds))
print('MAE :      ',mean_absolute_error(y_val,y_preds))
```

```python
svr.fit(x_train,y_train)
y_preds=svr.predict(x_val)
print('R2 Score : ',r2_score(y_val,y_preds))
print('MSE :      ',mean_squared_error(y_val,y_preds))
print('MAE :      ',mean_absolute_error(y_val,y_preds))
```

```python
dtr.fit(x_train,y_train)
y_preds=dtr.predict(x_val)
print('R2 Score : ',r2_score(y_val,y_preds))
print('MSE :      ',mean_squared_error(y_val,y_preds))
print('MAE :      ',mean_absolute_error(y_val,y_preds))
```

```python
rfr.fit(x_train,y_train)
y_preds=rfr.predict(x_val)
print('R2 Score : ',r2_score(y_val,y_preds))
print('MSE :      ',mean_squared_error(y_val,y_preds))
print('MAE :      ',mean_absolute_error(y_val,y_preds))
```

#### Tuning the models with better r2 score

```python
dtr= DecisionTreeRegressor(random_state=15)
kf=KFold(n_splits=5)
params={
    'criterion':['squared_error','absolute_error','friedman_mse','poisson'],
    'splitter':['best','random'],
    'max_depth':[3,4,5,6],
    'min_samples_split':[2,3,4,5],
    'min_samples_leaf':[1,2,3],
    'max_features':['sqrt','log2','auto']
}
gcvd=GridSearchCV(dtr,param_grid=params,cv=kf)
gcvd.fit(x_train,y_train)
y_preds=gcvd.predict(x_val)
print('Best Params',gcvd.best_params_,'\nBest Score',gcvd.best_score_,'\nBest Estimator',gcvd.best_estimator_)
print('R2 Score : ',r2_score(y_val,y_preds))
print('MSE :      ',mean_squared_error(y_val,y_preds))
print('MAE :      ',mean_absolute_error(y_val,y_preds))
```

### Features Selection Using RFE (with Tuned DecisionTreeRegressor )

```python
x_train.shape
```

```python
# The best number of features based on the lowest MSE score
from sklearn.feature_selection import RFE
estimator=gcvd.best_estimator_
results=[]
for num_feature in list(range(1,71)):
    rfe=RFE(estimator,n_features_to_select=num_feature)
    rfe.fit(x_train,y_train)
    gcvd.fit(x_train[:,rfe.support_],y_train)
    y_preds=gcvd.predict(x_val[:,rfe.support_])
    results.append(mean_squared_error(y_preds,y_val))

best_features_mse=(range(1, 71)[np.argmin(results)])
print('Best Number of Features: ',best_features_mse)
print('Best MSE Score: ',min(results),'\n')

rfe=RFE(estimator,n_features_to_select=best_features_mse)
rfe.fit(x_train,y_train)
gcvd.fit(x_train[:,rfe.support_],y_train)
y_preds=gcvd.predict(x_val[:,rfe.support_])
print('R2 Score : ',r2_score(y_val,y_preds))
print('MSE :      ',mean_squared_error(y_val,y_preds))
print('MAE :      ',mean_absolute_error(y_val,y_preds))
```

```python
print(f'Top {best_features_mse} Features with which model is tuned is : {df.drop("price", axis=1).columns[rfe.support_].values}')
```

```python
# The best number of features based on the lowest MAE score
results=[]
for num_feature in list(range(1,71)):
    rfe=RFE(estimator,n_features_to_select=num_feature)
    rfe.fit(x_train,y_train)
    gcvd.fit(x_train[:,rfe.support_],y_train)
    y_preds=gcvd.predict(x_val[:,rfe.support_])
    results.append(mean_absolute_error(y_preds,y_val))

best_features_mae=(range(1, 71)[np.argmin(results)])
print('Best Number of Features: ',best_features_mae)
print('Best MAE Score: ',min(results))

rfe=RFE(estimator,n_features_to_select=best_features_mae)
rfe.fit(x_train,y_train)
gcvd.fit(x_train[:,rfe.support_],y_train)
y_preds=gcvd.predict(x_val[:,rfe.support_])
print('R2 Score : ',r2_score(y_val,y_preds))
print('MSE :      ',mean_squared_error(y_val,y_preds))
print('MAE :      ',mean_absolute_error(y_val,y_preds))
```

```python
print(f'Top {best_features_mae} Features with which model is tuned is : {df.drop("price", axis=1).columns[rfe.support_].values}')
```

### Testing final_model on Test set

```python
# best_features_mse is obtained from best number of features based on the lowest MSE score 
rfe=RFE(estimator,n_features_to_select=best_features_mse)
rfe.fit(x_train,y_train)
gcvd.fit(x_train[:,rfe.support_],y_train)
y_preds=gcvd.predict(x_test[:,rfe.support_])
print('R2 Score : ',r2_score(y_test,y_preds))
print('MSE :      ',mean_squared_error(y_test,y_preds))
print('MAE :      ',mean_absolute_error(y_test,y_preds))
```
